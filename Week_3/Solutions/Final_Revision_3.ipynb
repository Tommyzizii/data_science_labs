```python
sales = pd.read_csv('sales_subset.csv',index_col=0)
sales.head()
```

```python
store_type = sales.drop_duplicates(subset=['store','type'])
store_type.head()
```

```python
store_dept = sales.drop_duplicates(subset=['store','department'])
store_dept.head(10000)
```

```python
is_holiday = sales[sales['is_holiday'] == True]

holiday_dates = is_holiday.drop_duplicates(subset=['date'])
holiday_dates['date'].head(10)
```

```python
count_type = store_type['type'].value_counts()
print(count_type)
count_type_nor = store_type['type'].value_counts(normalize=True)
print(count_type_nor)
```

```python
count_dept = store_dept['department'].value_counts().sort_index(ascending=False)
print(count_dept)
count_dept_nor = store_dept['department'].value_counts(normalize=True).sort_index(ascending=False)
print(count_dept_nor)
```

```python
sales_all = sales['weekly_sales'].sum()
print(sales_all)

salesA = sales[sales['type'] == 'A']['weekly_sales'].sum()
print(salesA)

salesB = sales[sales['type'] == 'B']['weekly_sales'].sum()
print(salesB)

salesC = sales[sales['type'] == 'C']['weekly_sales'].sum()
print(salesC)

sales_propn_by_type = [salesA, salesB, salesC] / sales_all
print(sales_propn_by_type)
```

```python
# For each store type, aggregate weekly_sales: get min, max, mean, and median

sales_stats = sales.groupby('type')['weekly_sales'].agg(['mean','median','min','max'])
print(sales_stats)
```

```python
# For each store type, aggregate unemployment and fuel_price_usd_per_l: get min, max, mean, and median

unemp_fuel_stats = sales.groupby('type')[['unemployment','fuel_price_usd_per_l']].agg(['mean','median','min','max'])
print(unemp_fuel_stats)
```

```python
avocados = pd.read_csv('avocados.csv')

avocados.head()
```

```python
# Get the total number of avocados sold of each size

nb_sold_by_size = avocados.groupby('size')['nb_sold'].sum()
print(nb_sold_by_size)

nb_sold_by_size.plot(kind='bar')
plt.show()
```

```python
# Get the total number of avocados sold on each date

nb_sold_by_date = avocados.groupby('date')['nb_sold'].sum()
print(nb_sold_by_date)

nb_sold_by_date.plot(kind='line',rot = 45)
plt.show()
```

```python
# Scatter plot of avg_price vs. nb_sold with title

avocados.plot(x='nb_sold',y='avg_price',kind='scatter',title='nbsold and avgprice')
plt.show()
```

```python
# Histogram of conventional avg_price 

avocados[avocados['type'] == 'conventional']['avg_price'].plot(kind='hist')
avocados[avocados['type'] == 'organic']['avg_price'].plot(kind='hist')
plt.legend(['conventional','organic'])
plt.show()
```
