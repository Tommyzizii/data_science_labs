```python
import pandas as pd
df = pd.read_csv("W2_Superstore_Sales.csv", index_col=0)
df['Order_Date'] = pd.to_datetime(df['Order_Date'],errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'],errors='coerce')
df['Order_Year'] = df['Order_Date'].dt.year
print(df.info())
```

```python

sales_by_years = df.groupby(["Order_Year"])['Sales'].sum()
print(sales_by_years)
```

```python
total_sales = df['Sales'].sum()
print(f"The Total sales between Year 2015-2018 is " + str(total_sales) + " USD")
```

```python
cat_by_orders = df.groupby("Category").count().reset_index().sort_values(['Sales'] , ascending=False)
highest_sales = cat_by_orders["Category"].values[0]
print(highest_sales)
```

```python
plt.bar(cat_by_orders["Category"],cat_by_orders["Sales"])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Products and Sales")
plt.show()
```

```python
cat_by_sales = df.groupby("Category")["Sales"].sum().reset_index().sort_values(["Sales"], ascending=False)
highest_sales = cat_by_sales["Category"].values[0]
print(highest_sales)
```

```python
plt.bar(cat_by_sales["Category"],cat_by_sales["Sales"])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Products and Sales")
plt.show()
```
