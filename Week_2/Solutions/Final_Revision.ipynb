```python
df = pd.read_csv('W2_Superstore_Sales.csv', index_col=0)
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], errors='coerce')
df['Order_Year'] = df['Order_Date'].dt.year
df.head()
```

```python
year_2015_2018 = (df['Order_Year'] >= 2015) & (df['Order_Year'] <= 2018)
result = df[year_2015_2018]['Sales'].sum()
print(result)
```

```python
salesByYear = df.groupby(['Order_Year'])['Sales'].sum()
salesByYear.plot(kind='bar')
plt.xlabel('Order by Year')
plt.ylabel('Sales Volume')
plt.title('Total sales by year (2015-2018)')
plt.show()
```

```python
salesByCat = df.groupby(['Category'])['Sales'].sum().reset_index().sort_values(['Sales'],ascending=True)
print(salesByCat)

salesByCat.plot(kind='bar')
plt.show()
```
