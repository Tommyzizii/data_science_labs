```python
df = pd.read_csv('W2_Superstore_Sales.csv',index_col=0)
df['Order_Date'] = pd.to_datetime(df['Order_Date'],errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'],errors='coerce')
df['Order_Year'] = df['Order_Date'].dt.year
df.head()

# change Date columns to usable DateTime in pandas, and add Order_Year column to the DataFrame. 
```

```python
sales = df['Sales']
total = 0.00
for i in sales:
    total += i

print(f"Total Sales between 2015-2018 is " + str(total) + " USD")
```

```python
sales_by_year = df.groupby(['Order_Year'])['Sales'].sum()
print(sales_by_year)
```

```python
sales_by_year.plot(kind='bar')
plt.xlabel('Order By Year')
plt.ylabel('Sales Volume')
plt.title('Total Sales By Year (2015-2018)')
plt.show()
```

```python
order_by_cat = df.groupby(['Category']).count().reset_index()
order_by_cat.head()
```

```python
plt.bar(order_by_cat['Category'],order_by_cat['Sales'])
plt.xlabel('Category')
plt.ylabel('Slaes')
plt.title('Sales By Category')
```

```python
sales_by_cat = df.groupby(['Category'])['Sales'].sum()
sales_by_cat.plot(kind='bar',ylabel='Sales',title='Sales by Category')
```

```python
sales_by_subcat = df.groupby(['Sub_Category'])['Sales'].sum().reset_index().sort_values(['Sales'],ascending=False)
sales_by_subcat.plot(kind='bar',xlabel='Sub-Category',ylabel='Sales Volume',title='Sales by Sub-Category')
```

```python
sales_by_region = df.groupby(['Region'])['Sales'].sum().reset_index().sort_values(['Sales'],ascending=False)
plt.bar(sales_by_region['Region'],sales_by_region['Sales'])
plt.xlabel('Regions')
plt.ylabel('Sales')
plt.title('Sales by Regions')
plt.show()
```
