```python
planes = pd.read_csv('Airlines_unclean.csv',index_col=0,parse_dates=['Date_of_Journey','Dep_Time','Arrival_Time'], date_format = "%d/%m/%Y")
planes.head()
```

```python
planes['Duration'] = planes['Duration'].str.replace('h','')
planes['Duration'] = planes['Duration'].str.replace('m','')
planes['Duration'] = planes['Duration'].str.replace(' ','.')
planes['Duration'] = planes['Duration'].astype(float)
planes['Duration'].head()
```

```python
print(planes.info())
```

```python
sns.heatmap(planes.corr(numeric_only=True),annot=True)
plt.show()

# To do this, we need to perform duration column first (From 00h 00m to float) with replace method
```

```python
ax = sns.heatmap(planes.corr(numeric_only=True), annot=True)
ax.set_ylim([0,2])
plt.show()
```

```python
print(planes['Total_Stops'].value_counts())
```

```python
threshold = len(planes)*0.05
print(threshold)
```

```python
print(planes.isna().sum())
```

```python
cols_to_drop = planes.columns[planes.isna().sum() <= threshold]
planes.dropna(subset=cols_to_drop,inplace=True)
```

```python
print(planes.isna().sum())
```

```python
planes.dropna(subset='Additional_Info',inplace=True)
```

```python
print(planes.isna().sum())
```

```python
airplane_prices = planes.groupby('Airline')['Price'].median()
print(airplane_prices)
```

```python
price_dict = airplane_prices.to_dict()
print(price_dict)
```

```python
planes['Price'] = planes['Price'].fillna(planes['Airline'].map(price_dict))
```

```python
planes.isna().sum()
```

```python
planes["Total_Stops"] = planes["Total_Stops"].str.replace(" Stops", "")
planes["Total_Stops"] = planes["Total_Stops"].str.replace(" Stop", "")
planes["Total_Stops"] = planes["Total_Stops"].str.replace("\\D", "",regex=True)
planes["Total_Stops"] = planes["Total_Stops"].astype(int)
```

```python
planes.head()
```

```python
sns.heatmap(planes.corr(numeric_only=True),annot=True)
plt.show()
```

```python
planes['month'] = planes['Date_of_Journey'].dt.month
planes['weekday'] = planes['Date_of_Journey'].dt.weekday

planes[['Date_of_Journey','month','weekday']].head()
```

```python
planes.head()
```

```python
planes['Dep_Hour'] = pd.to_datetime(planes['Dep_Time'],format='mixed')
planes['Arrival_Hour'] = pd.to_datetime(planes['Arrival_Time'],format='mixed')

planes.head()
```

```python
sns.heatmap(planes.corr(numeric_only=True),annot=True)
plt.show()
```

```python
salaries = pd.read_csv('Salaries_with_date_of_response.csv',index_col=0)
salaries.head()
```

```python
salaries['date_of_response'] = pd.to_datetime(salaries['date_of_response'],format='mixed')
```

```python
salaries.head()
```

```python
salaries['month'] = salaries['date_of_response'].dt.month
salaries.head()
```

```python
salaries['weekday'] = salaries['date_of_response'].dt.weekday
salaries.head()
```

```python
sns.heatmap(salaries.corr(numeric_only=True),annot=True)
plt.show()
```

```python
twenty_fifth = salaries['Salary_USD'].quantile(0.25)
median = salaries['Salary_USD'].median()
seventy_fifth = salaries['Salary_USD'].quantile(0.75)
maximum = salaries['Salary_USD'].max()
```

```python
salary_labels = ["entry","mid","senior","exec"]
salary_ranges = [0,twenty_fifth,median,seventy_fifth,maximum]
```

```python
salaries['salary_category'] = pd.cut(salaries['Salary_USD'],labels=salary_labels,bins=salary_ranges)
salaries[['Salary_USD','salary_category']].head()
```

```python
sns.countplot(data=salaries, x='Company_Size', hue='salary_category')
plt.show()
```
