```python
planes = pd.read_csv('Airlines_unclean.csv',index_col=0)
planes.head()
```

```python
non_numeric = planes.select_dtypes('object')
non_numeric.head()

for columns in non_numeric.columns:
    print(f"Number of unique values in {columns} columns", non_numeric[columns].nunique())
```

```python
duration_cateogry = [
    'short flight',
    'medium flight',
    'long flight'
]
```

```python
short_flight = '0h|1h|2h|3h|4h'
medium_flight = '5h|6h|7h|8h|9h'
long_flight = '10h|11h|12h|13h|14h|15h|16h'
```

```python
conditions = [(planes['Duration'].str.contains(short_flight)),
              (planes['Duration'].str.contains(medium_flight)),
              (planes['Duration'].str.contains(long_flight))]
```

```python
planes['Duration_Category'] = np.select(
    conditions,
    duration_cateogry,
    default='Other'
) 
```

```python
planes['Duration'].head()
```

```python
planes['Duration'] = planes['Duration'].str.replace('h','')
planes['Duration'] = planes['Duration'].str.replace('m','')
planes['Duration'] = planes['Duration'].str.replace(" ",".")
planes['Duration'].head()
```

```python
planes['Duration'] = planes['Duration'].astype(float)
planes['Duration'].head()
```

```python
sns.histplot(data=planes,x='Duration')
```

```python
planes['stdv_price'] = planes.groupby('Airline')['Price'].transform(lambda x: x.std())
planes.head()
```

```python
planes['airline_median_duration'] = planes.groupby('Airline')['Duration'].transform(lambda x: x.median())
planes.head()
```

```python
planes['price_destination_mean'] = planes.groupby('Destination')['Price'].transform(lambda x: x.mean())
planes.head()
```

```python
price_seventy_fifth = planes['Price'].quantile(0.75)
print("75 -> " + str(price_seventy_fifth))

price_twenty_fifth = planes['Price'].quantile(0.25)
print("25 -> " + str(price_twenty_fifth))

price_iqr = price_seventy_fifth - price_twenty_fifth
print("iqr -> " + str(price_iqr))

upper = price_seventy_fifth + (price_iqr * 1.5)
print('upper -> ' + str(upper))

lower = price_twenty_fifth - (price_iqr*1.5)
print('lower -> ' + str(lower))

no_outliners = planes[(planes['Price'] > lower)&(planes['Price'] < upper)]
no_outliners.describe()
```
